```python
import csv, json, codecs, uuid
import pandas as pd
import numpy as np
```

```python
datafile = 'Meteorite_Landings.csv'
```

## cleaning data file

```python
df = pd.read_csv(datafile,  encoding = "ISO-8859-1")
df
```

```python
dropped = df.dropna(axis=0, how='any')
dropped
```

```python
dropped = dropped[dropped.reclat.notnull()]
dropped = dropped[dropped.reclong.notnull()]
dropped
```

```python
dataset = dropped[0:100].drop(['id'], axis=1)
dataset
```

## saving to json

```python
# save to json
json_data = []
for row in dataset.iterrows():
    i, d = row
    json_data.append(d.tolist())
json_data
```

```python
headers = list(dataset)
```

```python
final_json = {}

headers = ['name', 'mass', 'year', 'reclat', 'reclong'] #manually adjust header names
final_json["name"] = "Meteor" #manually name json

final_json["id"] = str(uuid.uuid4())
final_json["classType"] = "imported"

initsig = {}
for header in headers:
    initsig[header] = []

final_json["initializedSignals"] = initsig
```

```python
#custom dict for signal info, manually change this too
signalDict = {
    'name' : {'displayName': 'Name', 'classType' : 'string', 'fieldName' : 'name'},
    'mass' : {'displayName': 'Mass (g)', 'classType' : 'number', 'fieldName' : 'mass'},
    'year' : {'displayName': 'Year', 'classType' : 'date', 'fieldName' : 'year'},
    'reclat' : {'displayName': 'Latitude', 'classType' : 'number', 'fieldName' : 'reclat'},
    'reclong' : {'displayName': 'Longitude', 'classType' : 'number', 'fieldName' : 'reclong'},
}
```

```python
signals = []
i = 0
for key, value in signalDict.items():
    signal = {}
    signal["id"] = str(uuid.uuid4())
    signal["order"] = i
    signal["style"] = "imported"
    signal["settings"] = {}
    
    signal["displayName"] = signalDict[key]["displayName"]
    signal["classType"] = signalDict[key]["classType"]
    signal["fieldName"] = signalDict[key]["fieldName"]
    
    i += 1
    signals.append(signal)
signals
```

```python
final_json["signals"] = signals
final_json["data"] = json_data

final_json
```

```python
with open ('meteor.json', 'w') as outfile:
    json.dump(final_json, outfile)
```
